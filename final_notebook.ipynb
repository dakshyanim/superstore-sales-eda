```python
plt.pie(state_distribution, labels=state_distribution.index, autopct='%1.1f%%')
plt.show()
```

```python
state_counts = df["State"].value_counts(normalize=True) * 100

# keep top 3
top_states = state_counts.head(3)

# combine rest as "Other"
other = pd.Series(
    {"Other": state_counts.iloc[3:].sum()}
)

# final data
final_data = pd.concat([top_states, other])

# pie chart
plt.pie(final_data, labels=final_data.index, autopct='%1.1f%%')
plt.title("Top States vs Others")
plt.show()
```

```python
df["City"].value_counts()
```

```python
city_counts = df["City"].value_counts(normalize=True) * 100

# keep top 3
top_city = city_counts.head(3)

# combine rest as "Other"
other = pd.Series(
    {"Other": city_counts.iloc[3:].sum()}
)

# final data
final_data = pd.concat([top_city, other])

# pie chart
plt.pie(final_data, labels=final_data.index, autopct='%1.1f%%')
plt.title("Top cities vs Others")
plt.show()
```

#LOYAL CUSTOMER

```python
top_customers = df["Customer Name"].value_counts().head(10).index

df[df["Customer Name"].isin(top_customers)] \
  .groupby("Customer Name")[["Sales", "Profit"]] \
  .sum() \
  .sort_values("Profit")
```

Insight:
Analysis of the most loyal customers (based on order frequency) reveals that loyalty does not always translate into profitability. While several frequent customers such as Paul Prost, Edward Hooks, and Matt Abelman contribute significantly to overall profit, others like Zuschuss Carroll result in net losses despite repeated purchases. This indicates that customer value varies widely among loyal customers and highlights the importance of evaluating both purchase frequency and profitability when assessing customer performance.
